```python
import requests
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Paso 1: Solicitud a la API
url = 'https://referenciales.cl/api/public/map-data'
response = requests.get(url)
data = response.json()

# Paso 2: Revisar si fue exitosa
if data.get('success'):
    df = pd.DataFrame(data['data'])
else:
    raise ValueError("No se pudo obtener la información")

# Paso 3: Crear geometría si existen columnas 'lat' y 'lng' (o similares)
df['geometry'] = df.apply(lambda row: Point(row['lng'], row['lat']), axis=1)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Paso 4: Graficar
gdf.plot(figsize=(10, 10), color='red', markersize=5)
plt.title("Puntos obtenidos desde la API de Referenciales.cl")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.show()
```

```python
import requests
import pandas as pd
import folium

# Obtener datos desde la API
url = 'https://referenciales.cl/api/public/map-data'
response = requests.get(url)
data = response.json()

# Validar y cargar al DataFrame
if data.get('success'):
    df = pd.DataFrame(data['data'])
else:
    raise ValueError("No se pudieron obtener los datos")

# Asegúrate de que las columnas se llamen 'lat' y 'lng' o edita aquí
lat_col = 'lat'
lng_col = 'lng'

# Crear mapa centrado en la media de los puntos
center_lat = df[lat_col].mean()
center_lng = df[lng_col].mean()

m = folium.Map(location=[center_lat, center_lng], zoom_start=5)

# Agregar marcadores
for _, row in df.iterrows():
    folium.Marker(
        location=[row[lat_col], row[lng_col]],
        popup=str(row.get('numero', 'Punto sin nombre'))
    ).add_to(m)

# Mostrar el mapa
m
```
